```python
#splitting the data
```

```python
X=df.drop('class',axis=1)
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
#training the data
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

```python
cols=['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium',
       'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
       'Proanthocyanins', 'Color intensity', 'Hue',
       'OD280/OD315 of diluted wines', 'Proline']
```

```python
accuracy_list=[]
```

```python
def model(model,data,accuracy_list):
    
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    
    print(classification_report(y_test,y_pred))
    print('\n')
    accuracy=accuracy_score(y_test,y_pred).round(2)
    accuracy_list.append(accuracy)
    print(f"The accuracy is {accuracy}")
    print('\n')
    
    confusion_matrix(y_test,y_pred)
    cm=confusion_matrix(y_test,y_pred)
    df_cm=pd.DataFrame(cm,index=[i for i in ['1','2','3']],
                      columns=[i for i in [['Predicted 1','Predicted 2','Predicted 3']]])
    plt.figure(figsize=(7,4))
    sns.heatmap(data=df_cm,annot=True,fmt='g')
```

```python
#Logistic Regression
```

```python
lg=LogisticRegression()
model(lg,df,accuracy_list)
```

```python
#Gaussian NB
```

```python
gb=GaussianNB()
model(gb,df,accuracy_list)
```

```python
#Decision Tree
```

```python
dt=DecisionTreeClassifier()
model(dt,df,accuracy_list)
```

```python
#KNN
```

```python
knn=KNeighborsClassifier()
model(knn,df,accuracy_list)
```

```python
#Random Forest
```

```python
rf=RandomForestClassifier()
model(rf,df,accuracy_list)
```

```python
#Adaboost
```

```python
ad=AdaBoostClassifier()
model(ad,df,accuracy_list)
```

```python
#Gradient Boost
```

```python
gb=GradientBoostingClassifier()
model(gb,df,accuracy_list)
```

```python
#Support Vector
```

```python
svm=SVC()
model(svm,df,accuracy_list)
```

```python
accuracy_list
```

```python
accuracies={'Logistic Regression':accuracy_list[0],
           'Gaussian NB':accuracy_list[1],
           'Decision Tree':accuracy_list[2],
           'KNN':accuracy_list[3],
           'Random Forest':accuracy_list[4],
           'AdaBoost':accuracy_list[5],
           'Gradient Boost':accuracy_list[6],
           'Support Vector':accuracy_list[7]}
```

```python
final_model=RandomForestClassifier()
model(final_model,df,accuracy_list)
```
