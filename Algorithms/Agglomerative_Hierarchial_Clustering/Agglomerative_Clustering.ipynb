```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python
df.head()
```

```python
#we wil use only the 3rd and 4th columns as our data
```

```python
data=df.iloc[:,3:5].values
```

```python
data
```

```python
#we will now plot the dendogram for this data using scipy library
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10,10))
plt.title("Customer Dendograms")
dend=shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
#now we build an agglomerative clustering model
```

```python
from sklearn.cluster import AgglomerativeClustering

ag=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
labels=ag.fit_predict(data)
```

```python
labels
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(data[:, 0], data[:, 1], c=ag.labels_, cmap='rainbow')
```
