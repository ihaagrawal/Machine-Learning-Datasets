```python
#splitting the model
```

```python
X=df.drop('species',axis=1)
y=df['species']
```

```python
X
```

```python
y
```

```python
#train and test split
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#training the model
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
#fiiting and predicting 
```

```python
def model(model,data):
    
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    
    print(classification_report(y_test,y_pred))
    print('\n')
    print(f"The accuracy is {(accuracy_score(y_test,y_pred).round(2))}")
    print('\n')
    
    confusion_matrix(y_test,y_pred)
    cm=confusion_matrix(y_test,y_pred,labels=['Iris-setosa','Iris-versicolor','Iris-virginica'])
    df_cm=pd.DataFrame(cm,index=[i for i in ['Iris-Setosa','Iris-Versicolor','Iris-Virginica']],
                      columns=[i for i in [['Predicted Iris-Setosa','Predicted Iris-Versicolor','Predicted Iris-Virginica']]])
    plt.figure(figsize=(7,4))
    sns.heatmap(data=df_cm,annot=True,fmt='g')
    
```

```python
#Logistic Regression
```

```python
lg=LogisticRegression()
model(lg,df)
```

```python
#GaussianNB
```

```python
gb=GaussianNB()
model(gb,df)
```

```python
#Decision Tree
```

```python
dt=DecisionTreeClassifier()
model(dt,df)
```

```python
#KNeighbors
```

```python
kn=KNeighborsClassifier()
model(kn,df)
```

```python
#Random Forest
```

```python
rf=RandomForestClassifier()
model(rf,df)
```

```python
#Gradient Boosting
```

```python
gb=GradientBoostingClassifier()
model(gb,df)
```

```python
#Adaboost
```

```python
ab=AdaBoostClassifier()
model(ab,df)
```

```python
#Support Vector Machine
```

```python
svm=SVC()
model(svm,df)
```

```python
#All the algorithms except Gaussian NB give 100% accuracy, Gaussian NB gives 98% accuracy
```

```python
#finalising
```

```python
#we will use Random Forest for the final model
```

```python
final_model=RandomForestClassifier()
model(final_model,df)
```
