```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import fetch_20newsgroups
df=fetch_20newsgroups()
```

```python
df.target_names
```

```python
cat=['alt.atheism',
 'comp.graphics',
 'comp.os.ms-windows.misc',
 'comp.sys.ibm.pc.hardware',
 'comp.sys.mac.hardware',
 'comp.windows.x',
 'misc.forsale',
 'rec.autos',
 'rec.motorcycles',
 'rec.sport.baseball',
 'rec.sport.hockey',
 'sci.crypt',
 'sci.electronics',
 'sci.med',
 'sci.space',
 'soc.religion.christian',
 'talk.politics.guns',
 'talk.politics.mideast',
 'talk.politics.misc',
 'talk.religion.misc']
train=fetch_20newsgroups(subset='train', categories=cat)
test=fetch_20newsgroups(subset='test', categories=cat)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.naive_bayes import MultinomialNB
```

```python
model=make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
model.fit(train.data, train.target)
```

```python
labels=model.predict(test.data)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
score=accuracy_score(test.target, labels)
cm=confusion_matrix(test.target, labels)
score
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(cm, annot=True, fmt='0.2f', xticklabels=cat, yticklabels=cat)
plt.xlabel('true label')
plt.ylabel('predicted label')
```

```python
def predict_category(s, train=train, model=model):
    pred=model.predict([s])
    return train.target_names[pred[0]]
```

```python
predict_category('Jesus christ')
```
